```python
# using image data generator
data_gen = ImageDataGenerator(rescale = 1.0/255,validation_split = 0.2 )# 20 percent is validation dataset
```

```python
train_gen = data_gen.flow_from_directory(base_dir,target_size =(224,224),batch_size=64,subset='training',class_mode = 'categorical')
val_gen = data_gen.flow_from_directory(base_dir,target_size =(224,224),batch_size=64,subset='validation',class_mode = 'categorical')
```

# Building the CNN

```python
# Model Definition
no_of_classes = 38
model = Sequential()
model.add(layers.Conv2D(32,3,activation='relu',input_shape=(224,224,3)))# 32 filter of 3x3 size
model.add(layers.MaxPooling2D(2,2)) #  pool_size = stride = (2,2)
model.add(layers.Conv2D(64,3,activation='relu'))# 64 filter of 3x3 size
model.add(layers.MaxPooling2D(2,2)) #  pool_size = stride = (2,2)
model.add(layers.Flatten())
model.add(layers.Dense(128,activation='relu'))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(no_of_classes,activation="softmax"))

model.summary()
```

```python
# compile the model
batch_size = 32
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Training the Model
history = model.fit(
    train_gen,
    steps_per_epoch=train_gen.samples // batch_size,  # Number of steps per epoch
    epochs=5,  # Number of epochs
    validation_data=val_gen,
    validation_steps=val_gen.samples // batch_size  # Validation steps
)
```

```python
# evaluate the model
loss,accuracy = model.evaluate(val_gen,steps=val_gen.samples // batch_size)
print(f"Loss: {loss}")
print(f"Accuracy: {accuracy}")
```

```python
plt.plot(history.history['accuracy'],label='train')
plt.plot(history.history['val_accuracy'],label='validation')
plt.title("Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("accuracy")
plt.legend(loc='upper left')
plt.show()
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='validation')
plt.title("Model Loss")
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.legend(loc='upper left')
plt.show()
```

```python
# Building predictive model

from matplotlib import image as mpimg
from PIL import Image
def predict_class(model,path,class_indices):
   img1 = Image.open(path)
   img1 = img1.resize((224,224))
   img1 = img1.convert('RGB')
   img1 = np.array(img1)
   img1 = img1 / 255.0  # optional depending on model preprocessing

# Add batch dimension ---> important step (1,224,224,3)
   img1 = np.expand_dims(img1, axis=0)
# Predict
   y_p = model.predict(img1)
   y_predicted = y_p.argmax(axis=1)
   return  class_indices[y_predicted[0]]
```
